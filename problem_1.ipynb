## Problem 1: One die with S sides

### 1. What are the theoretical values of the mean, variance, and standard deviation as functions of S? The answers should be in closed form, rather than expressed as a sum.

#### mean: &emsp; $\langle s\rangle = \frac{(s+1)}{2}$ <br>
#### variance: &emsp; $\sigma^2 = \frac{(s-1)(s+1)}{12}$ <br>
#### standard: &emsp; $\sigma = \sqrt{\frac{(s-1)(s+1)}{12}}$

### 2. Modify the program you used for an earlier problem to simulate rolling a die with S sides, and print out the mean, variance, and standard deviation for a number of trials. Have the program print out the theoretical predictions in each case, as well as the deviations from theory to facilitate comparisons.

my code:

```python
import time
import numpy as np

def OneDieS_np(trials, s):
    c1=time.clock()
    print("=================================================================================")
    print("Number of sides = ", s)
    print("Number of trials = ", trials)
    mean_th=(s + 1)/2
    variance_th=(s - 1)*(s + 1)/12
    standard_th=variance_th**0.5
    

    histogram = np.zeros(s)
    print(histogram)

    r = 0
    for t in range(trials):
        r = int(np.random.random()*s)
        histogram[r] = histogram[r] + 1
    print(histogram)
    pro = histogram/trials
    side = [i+1 for i in range(s)]
    mean_ex = np.dot(side,pro)
    sidee = [(i+1)**2 for i in range(s)]
    var = np.dot(sidee,pro)
    vari = (np.dot(side,pro))**2
    variance_ex = var-vari
    standard_ex = variance_ex**0.5
    print('mean:', mean_ex)
    print('variance:',variance_ex)
    print('standard:',standard_ex)
    
    
    print("------------------------------------------------------------------------------")
    
    print("| Mean-th | Mean-ex | Variance-th | Variance-ex | Standard-th | Standard-ex |")
    
    print("|{:^9.3f}|{:^9.3f}|{:^13.3f}|{:^13.3f}|{:^13.3f}|{:^13.3f}|".format(
    mean_th, mean_ex, variance_th, variance_ex, standard_th, standard_ex))
    
    print("------------------------------------------------------------------------------")
    
    print("Mean error : {:2.2f}%".format(100 * abs(mean_th-mean_ex) / mean_th))
    print("Variance error : {:2.2f}%".format(100 * abs(variance_th-variance_ex) / variance_th))
    print("Standard error : {:2.2f}%".format(100 * abs(standard_th-standard_ex) / standard_th))
    

    c2=time.clock()
    print("Elapsed time =", c2-c1)

def bobo():
    OneDieS_np(10000, 8)
    OneDieS_np(1000000, 8)
bobo()
print("\n{:>70}".format("By Bobo"))
```
